## Plot Rainfall Intensity - Duration - Frequency Curves (IDF Curve)

```python
#Import modul require
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import dataset
df = pd.read_csv("rainfall_idf_curves_for_delaware.csv")
df
```

```python
# Plot Tc of Return Period
ax = df.plot(x="Tc_(min)",
             y="2_YEARS",
             xlabel="Time of Concentration (min)",
             ylabel="Rainfall Intensity (in/hr)",
             label="2-years",
             title="Rainfall Intensity - Duration - Frequency Curves \n Sussex County, Delaware"
            )
df.plot(x="Tc_(min)",
        y="5_YEARS",
        label="5-years",
        ax=ax,
        )

df.plot(x="Tc_(min)",
        y="10_YEARS",
        label="10-years",
        ax=ax,
        )

df.plot(x="Tc_(min)",
        y="25_YEARS",
        label="25-years",
        ax=ax,
        )

df.plot(x="Tc_(min)",
        y='50_YEARS',
        label="50-years",
        ax=ax,
        )

df.plot(x="Tc_(min)",
        y="100_YEARS",
        label="100-years",
        ax=ax,
        )

# Showing the grid
plt.grid()
```

```python
fig, ax = plt.subplots()
ax.plot(df["Tc_(min)"], df["2_YEARS"], label=("2-years"))
ax.set_xlabel('Time of Concentration (min)')
ax.set_ylabel('Rainfall Intensity (in/hr)')
ax.set_title("Rainfall Intensity - Duration - Frequency Curves \n Sussex County, Delaware")
ax2 = ax
ax2.plot(df["Tc_(min)"], df["5_YEARS"], label=("5-years"))
ax2 = ax
ax2.plot(df["Tc_(min)"], df["10_YEARS"], label=("10-years"))
ax2 = ax
ax2.plot(df["Tc_(min)"], df["25_YEARS"], label=("25-years"))
ax2 = ax
ax2.plot(df["Tc_(min)"], df["50_YEARS"], label=("50-years"))
ax2 = ax
ax2.plot(df["Tc_(min)"], df["100_YEARS"], label=("100-years"))

# Showing the grid and legend
plt.grid()
ax.legend()

# Showing the plot
plt.show()
```

```python
def plot_idf(axes, x, y, color, dashes, label):
  axes.plot(x, y, color=color, dashes=dashes, label=label)
    
# Plot with def
fig, ax = plt.subplots(figsize=[5.35, 7.25])
plot_idf(ax, df["Tc_(min)"], df["2_YEARS"], 'black', [0, 3, 3, 1, 3, 1, 3], "2-years")
ax2 = ax
plot_idf(ax2, df["Tc_(min)"], df["5_YEARS"], 'black', [0, 1, 1], "5-years")
ax2 = ax
plot_idf(ax2, df["Tc_(min)"], df["10_YEARS"], 'black', [10], "10-years")
ax2 = ax
plot_idf(ax2, df["Tc_(min)"], df["25_YEARS"], 'black', [0, 3, 5, 1, 5], "25-years")
ax2 = ax
plot_idf(ax2, df["Tc_(min)"], df["50_YEARS"], 'black', [6, 2], "50-years")
ax2 = ax
plot_idf(ax2, df["Tc_(min)"], df["100_YEARS"], 'black', [1], "100-years")

# Setting the x and y labels
ax.set_xlabel('Time of Concentration (min)')
ax.set_ylabel('Rainfall Intensity (in/hr)')

# Setting the title
ax.set_title("Rainfall Intensity - Duration - Frequency Curves \n Sussex County, Delaware")

# Setting the x and y limit
ax.set_xlim((0, 180))
ax.set_ylim((0, 9))

# Showing the grid and legend
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3, fancybox=True)
plt.grid()

# Showing the plot
plt.show()
```
